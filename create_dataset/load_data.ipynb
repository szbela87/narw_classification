```python
from scipy.io import arff
import pandas as pd
from sklearn.model_selection import train_test_split
import os
```

```python
#os.chdir("/home/szekeres/Dokumentumok/RightWhaleCalls")
!pwd
```

```python
SEED = 2023
```

```python
# Load ARFF files
data_train, meta = arff.loadarff('RightWhaleCalls_TRAIN.arff')
data_test, meta = arff.loadarff('RightWhaleCalls_TEST.arff')
```

```python
#meta
```

```python
# Convert to a Pandas DataFrame if needed
df_train = pd.DataFrame(data_train)
df_test = pd.DataFrame(data_test)
```

```python
df_train.head()
```

```python
targets = df_train['class'].unique()
targets
```

```python
target_values_dict = {}
for i, name in enumerate(targets):
    target_values_dict[name] = i
```

```python
target_values_dict
```

```python
df_train["class"] = df_train["class"].replace(target_values_dict)
df_test["class"] = df_test["class"].replace(target_values_dict)
df_train.head()
```

```python
len(df_train), len(df_test)
```

```python
df = pd.concat([df_train, df_test], ignore_index=True)
```

```python
len(df)
```

```python
df_train, df_test = train_test_split(df, stratify = df["class"], test_size = 0.2, random_state = SEED)
```

```python
df_train
```

```python
df_train.to_csv("train_whales.csv",sep=",",index=False)
df_test.to_csv("test_whales.csv",sep=",",index=False)
```
